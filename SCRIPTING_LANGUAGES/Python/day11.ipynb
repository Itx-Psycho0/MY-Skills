```python
import numpy as np

# Columns: [Age, Math Marks, Science Marks]
data = np.array([
    [18, 85, 78],   # Student 1
    [19, 92, 88],   # Student 2
    [17, 76, 95],   # Student 3
    [18, 65, 70],   # Student 4
    [20, 90, 85]    # Student 5
])
```

```python
# Get the shape of the matrix.
data.shape
```

```python
# find avg age of std.
np.mean(data[:,0],axis=0)
```

```python
# Extract Math marks of all students
data[:,1]
```

```python
# Find the highest Science mark.
data[:,2].max()
```

```python
# Get details of the std who scored more than 90 in math
data[data[:,1]>90]
data
```

```python
# Increase Math marks of all std by 5
data[:,1]+5
data
```

```python
# find how many std are younger than 19
len(data[data[:,0]<19])
```

```python
# Calculate the avg marks in each sub(column-wise mean)
np.mean(data[:,1:],axis=0)
```

```python
# get data of std who scored at least 80 in both sub

data[data[:,1]>=80] & (data[data[:,2]>=80])
```
