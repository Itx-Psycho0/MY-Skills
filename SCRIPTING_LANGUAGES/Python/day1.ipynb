```python
# In Python, we communicate with the user (or ourselves) using a function called print().
print('Hello, World!')

# print: This is the command (function).

# (): The parentheses tell Python to execute the function.

# "Hello, World!": This is the data (a String) we want to show. Quotes are required for text
```

What do you think would happen if we removed the quotation marks "" around Hello, World?

If you write print(Hello) without quotes, Python will give you an error (specifically a NameError).

Why? Computers take everything literally. To speak to Python, you need to know what type of data you are handing it. This is the concept of Data Types.

Here are the three most fundamental types you must memorize today:

1. Strings (Text)
To Python, text is just a string of characters. To tell Python "This is text, not a command," you must wrap it in quotation marks (either single ' or double ").

Code: "Hello World" or 'Python is cool'

Analogy: Think of the quotes like an envelope. The message is inside, but the envelope keeps it safe so Python knows it's just a message to be carried, not an instruction to follow.

2. Integers (Whole Numbers)
These are whole numbers used for math. They do not use quotes.

Code: 5, 100, -20

Rule: If you put quotes around a number like "5", Python treats it as text (a String), not math!

3. Floats (Decimal Numbers)
These are numbers with decimal points.

Code: 3.14, 9.8, 0.001

```python
# we are going to write the program that process dofferent types

print('Hello')
print(3)
print(3.3)
print(3+2)
print("10"+"10")
# print(5 + "10") it's gimme the error --> TypeError: unsupported operand type(s) for +: 'int' and 'str'
```

the full definition (signature) of the print function looks like this: 

--- print(*objects, sep=' ', end='\n', file=sys.stdout, flush=False)----

1. *objects (The Data)
This is what you want to print. The * symbol means it accepts variable arguments—you can pass as many items as you want, separated by commas.

Example: print("A", "B", "C")

Behavior: Python converts all items to strings and displays them.

------------------------------------------------------------------------------------------------------------

2. sep (Separator)
"What goes between the items?"

Default: A single space ' '.

Customization: You can change this to anything—a comma, a dash, or even nothing at all.

Python

Default behavior
print("Apple", "Banana", "Cherry") 
Output: Apple Banana Cherry

Custom separator
print("Apple", "Banana", "Cherry", sep="-") 
Output: Apple-Banana-Cherry

No separator (useful for building single words from parts)
print("Super", "man", sep="") 
Output: Superman

-------------------------------------------------------------------------------------------------------------
3. end (End Character)
"What happens after the printing is done?"

Default: A newline character \n (this is why print usually jumps to the next line).

Customization: You can keep the cursor on the same line by changing this to a space or empty string.

Python

Default: jumps to next line
print("Hello")
print("World")
Output:
Hello
World

Custom: stays on same line
print("Hello", end=" ")
print("World")
Output: Hello World

-------------------------------------------------------------------------------------------------------------

4. file (Output Destination)
"Where should the text go?"

Default: sys.stdout (Standard Output, which is your screen/terminal).

Customization: You can send the text directly into a text file on your computer!

Python

Open a file named 'log.txt' in 'write' mode

with open('log.txt', 'w') as f:
    # This text will NOT appear on screen, but inside the file!
    print("This is a secret message.", file=f)


-------------------------------------------------------------------------------------------------------------


5. flush (Force Output)
"Should I display this immediately?"

Default: False (Python buffers output, meaning it waits to gather a chunk of text before displaying it to be more efficient).

Customization: Set to True to force text to appear instantly. This is crucial for things like loading bars or countdown timers where you want to see updates one by one, real-time.

```python
# Based on what you just learned, look at this code. What will be the exact output?

print("Python","JavaScript","RUBY",sep='|',end="-->")
print("Scripting Language")

#Option A: Python | Java | RUBY --> Scripting Language

#Option B: Python | Java | RUBY -->
#  Scripting Language

#Option C: Python Java RUBY --> Scripting Language

```
