# DataFrame

-> When you combine multiple series together

```python
# Creating a dataframe
data = {
    'Name':["john","Anna","peter","linda"],
    'Age' :[28,34,29,42],
    'city' : ["New York", "Paris", "Berlin", "London"],
    'Salary' : [65000, 7000, 6200, 85000]
}
df = pd.DataFrame(data)
```

```python
data_list = [
    ['John', 28, 'New York', 65000],
    ['Anna', 34, 'Paris', 70000],
    ['Peter', 29, 'Berlin', 62000],
    ['Linda', 42, 'London', 85000]
]
df2 = pd.DataFrame(data_list)
df2
columns = ["Name", "Age", "City", "Salary"]
df2 = pd.DataFrame(data_list,columns=columns)
df2
```

```python
# Selection and  Indexing of Columns

df2['Name']
df2[['Name','City']]
```

```python
# Creating a new Column
df2["Desi"] = ['dr','en','dr','en']
df2
```

```python
# Removing Columns
df2.drop('Desi',axis = 1) # its only saw copy of the dataframe not original
```

```python
df2
```

```python
df2.drop('Desi',axis = 1, inplace=True) # now its mutate original one
```
