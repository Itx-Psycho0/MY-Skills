```python
# Matplotlib
x = np.linspace(0,5,11)
x
y = x**3
y
```

```python
plt.plot(x,y) #plot a line graph
plt.title("Line graph")
plt.xlabel("ek")
plt.ylabel("teen")
```

```python
# SubPLots
plt.subplot(2,2,1)
plt.plot(y**2,x) 
plt.subplot(2,2,4)
plt.plot(x,y)
plt.subplot(2,2,3)
plt.plot(y,x) 
plt.subplot(2,2,2)
plt.plot(x**2,x**2)
```

```python
#Object oriented way
fig = plt.figure(figsize=(16,9),dpi = 500)
a1 =fig.add_axes([0.1,0.1,0.8,0.8]) # [margin top, margin bottom,width, height]
a1.plot(x,y)
a2 =fig.add_axes([0.5,0.5,0.4,0.4])
a2.plot(x,y)
```

```python
# Types of plots
plt.scatter(x,y)
```

```python
from random import sample
data = sample(range(1,1000),100)
plt.hist(data)
```

```python
plt.boxplot(x)
```

```python
# Saving plots
fig = plt.figure()
ax =fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x,y)
plt.savefig("basicplot.png")
```
