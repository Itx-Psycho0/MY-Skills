```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
ef = sns.load_dataset('tips')
ef
```

```python
ef['size'].unique()
```

```python
plt.subplot(1,2,1)

sns.histplot(ef['total_bill'],bins=20,kde = True)
plt.subplot(1,2,2)
sns.histplot(ef['tip'],kde=True)
```

```python
# joint plot
sns.jointplot(x = 'total_bill', y= 'tip', data = ef, kind='scatter')
```

```python
sns.jointplot(x = 'total_bill', y= 'tip', data = ef, kind='reg')
```

```python
# pair plot
sns.pairplot(ef, hue='sex')
```

```python
#rugplot
sns.rugplot(ef)
```

```python
# Categorial plots
# use for qualtiative data

#Count plot
sns.countplot(x=ef['sex'],hue = ef['smoker'])
```

```python
import numpy as np



# bar plot

sns.barplot(x = ef['sex'], y= ef['total_bill'],estimator=np.sum)
```

```python
#boxplot
sns.boxplot(x = 'tip', y = 'day', data = ef , palette='rainbow')
```

```python
sns.violinplot(x = ef['day'], y = 'total_bill', data = ef)
```

```python
# Strip plot and swarmplot 
sns.stripplot(x = 'day', y = 'total_bill', data = ef)
```

```python
sns.swarmplot(x = 'day', y = 'total_bill', data = ef)
sns.violinplot(x = ef['day'], y = 'total_bill', data = ef)
```
