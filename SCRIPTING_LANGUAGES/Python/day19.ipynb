```python
import matplotlib.pyplot as plt

def generate_movie_chart():
    # STEP 1: Define your data
    # TODO: Create a list named 'genres' with: Sci-Fi, Comedy, Drama, Action, Horror
    genres = ['Sci-Fi','Comedy','Drama',"Action","Horror"]

    # TODO: Create a list named 'votes' with: 45, 30, 20, 50, 5
    votes = [45,30,20,50,5]


    # STEP 2: Create the plot
    # TODO: Use the horizontal bar function: plt.barh(y_axis, x_axis)
    # Use the color code '#E74C3C' and add a black edgecolor
    plt.barh(genres, votes, color='#E74C3C', edgecolor='black')


    # STEP 3: Add the context
    # TODO: Set the title to 'Most Popular Movie Genres'
    plt.title('Most Popular Movie Genres')

    # TODO: Label the X-axis as 'Number of Votes' and Y-axis as 'Genres'
    plt.xlabel('Number of Votes')
    plt.ylabel('Genres')


    # STEP 4: Export and Display
    # TODO: Save the figure as 'movie_survey.png' with dpi=300
    plt.savefig('movie_survey.png', dpi=300, bbox_inches='tight')


    plt.show()
    return plt.gca()

# This ensures the chart still generates if you run main.py directly
if __name__ == "__main__":
    generate_movie_chart()
```

```python
def analyze_fitness_data():
    # 1. Prepare the data
    # Create a list called 'days' with values: 'Day 1' through 'Day 6'
    # Create a list called 'distance' with values: 5, 7, 6, 9, 12, 11
    days = ["Day 1", "Day 2", "Day 3", "Day 4", "Day 5", "Day 6"]
    distance = [5,7,6,9,12,11]


    # 2. Create the line chart
    # Use plt.plot() with: purple color, square markers ('s'), and a linewidth of 3
    plt.plot(days,distance,color='purple', marker='s',linewidth=3)

    # 3. Add Context
    # Add the title: "Daily Running Progress"
    # Add the X-axis label: "Training Day"
    # Add the Y-axis label: "Distance (km)"
    
    plt.title('Daily Running Progres')
    # plt.title('Weekly Temperature Trend', fontsize=14, fontweight='bold')
    plt.xlabel('Training Day')
    plt.ylabel('Distance (km)')

    # 4. Add Formatting
    # Add a light grid (use linestyle=':' and alpha=0.5)
    plt.grid(linestyle=':', alpha=0.5)


    # 5. Save and Show
    # Save the figure as 'fitness_progress.png'
    plt.savefig('fitness_progress.png')
    # Finally, display the plot
    plt.show()

    return plt.gca()

if __name__ == "__main__":
    analyze_fitness_data()
```

```python


def generate_language_chart():
    # 1. Data preparation
    langs = ['Hindi', 'English', 'Kannada', 'French']
    votes = [10, 2, 7, 5]

    # 2. Plot the pie chart
    plt.pie(votes, labels=langs, autopct='%1.1f%%', shadow=True, startangle=90)

    # 3. Ensure perfect circle
    plt.axis('equal')

    # 4. Add title
    plt.title('Programming Language Popularity')

    # 5. Save and display
    plt.savefig('language_popularity.png', dpi=300, bbox_inches='tight')
    plt.show()

if __name__ == "__main__":
    generate_language_chart()
```

```python
import matplotlib.pyplot as plt

def analyze_study_performance():
    
    # 1. Prepare the Data
    # Create a list 'hours' with the study times
    # Create a list 'scores' with the test results
    hours=[1,2,3,4,5,6,7,8]
    scores= [50,55,62,65,75,82,88,95]
    # 2. Create the Scatter Plot
    # Use plt.scatter() with: green color, size (s) of 120, and alpha of 0.6
    plt.scatter(hours,scores,s=120,c="green",alpha=0.6)

    # 3. Add Context
    # Add the title: "Impact of Study Time on Test Scores"
    # Label the X-axis: "Hours Spent Studying"
    # Label the Y-axis: "Score (%)"
    plt.title("Impact of Study Time on Test Scores")
    plt.xlabel("Hours Spent Studying")
    plt.ylabel("Score (%)")


    # 4. Save and Show
    plt.savefig('study_performance.png')
    plt.show()
    # Save the figure as 'study_performance.png'

    return plt.gca()

if __name__ == "__main__":
    analyze_study_performance()
```
