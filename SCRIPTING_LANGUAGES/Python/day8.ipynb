```python
# Numpy Arrays
# - [1,2,3,4] - array and list but what diff.

# list - datatype - mixed , performance - slower, Memory Efficiency - Low, Vector operations - Manual loops.

# Numpy array - datatype - Same datatype only, performance - Faster(c-level), Memory Efficiency - High, vector operations - Easy Fast.

import numpy as np

# Basic Array creation
arr = np.array([1,2,3,4])
print(arr)

a = [1,2,3,4,5,'hello']

print(np.array(a))
# it will shows as array(['1' '2' '3' '4' '5' 'hello'], dtype='<U32')


# Vectors
# single dimensional

# Matrix
# 2dimensional

# tensor
# more than 2d

l = [[1,2,3,5],[6,7,8,9],[10,11,12,13]]
print(l)

print(np.array(l))
```

```python
# array generation fn
# import numpy as np

#arange()
arr = np.arange(1,22)
print(arr)


arr2 = np.arange(1,20,2)
print(arr2)

#zeros()

arr3 = np.zeros(6)
print(arr3)
arr4 = np.zeros((6,6))
#.zeros((rows,column))
print(arr4)

# ones()

arr5 = np.ones((4,8))
print(arr5)

# linspace
arr6 = np.linspace(1,5,3) # means linear space
print(arr6)

print(arr)
```

```python
# Random Generation fn

# rand (normalization)
arr = np.random.rand(5)
print(arr)

# randn (standardization)
arr2 = np.random.randn(10)
print(arr2)

# randint
arr3 = np.random.randint(10,20,10)
print(arr3)
```
