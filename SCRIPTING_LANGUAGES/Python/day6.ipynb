# --> LIST <--
- Ordered collection of items
- Mutable (can be changed)
- Allows duplicate elements
- Defined using square brackets []
- Example:
  ```python
  my_list = [1, 2, 3, 4, 5]
  my_list.append(6)  # Adding an element
  print(my_list)     # Output: [1, 2, 3, 4, 5, 6]
  ```

````python
# Operations on lists
my_list = [10, 20, 30, 40, 50]
my_list.append(60)  # Add an element to the end
my_list.remove(30)  # Remove an element
print(my_list)  # Output the modified list
print(my_list[2])  # Access an element by index
my_list.sort()  # Sort the list
print(my_list)
my_list.reverse()  # Reverse the list
print(my_list)
my_list.insert(2, 25)  # Insert an element at index 2
print(my_list)
my_list.pop()  # Remove and return the last element
print(my_list)
my_list.extend([70, 80])  # Extend the list with another list
print(my_list)
my_list.clear()  # Clear the list
print(my_list)
my_list = [10, 20, 30, 40, 50]  # Reinitialize the list
print(len(my_list))  # Get the length of the list
print(my_list.count(20))  # Count occurrences of an element
print(my_list.index(40))  # Find the index of an element
my_list_copy = my_list.copy()  # Create a shallow copy of the list
print(my_list_copy)
my_list.sort(reverse=True)  # Sort the list in descending order
print(my_list)
del my_list  # Delete the list entirely
# print(my_list)  # This will raise an error since the list is deleted

# question: Write a Python program to demonstrate various operations on lists including adding, removing, accessing elements, sorting, reversing, inserting, popping, extending, clearing, counting occurrences, finding index, copying, and deleting the list.
# solution: The code above demonstrates all the requested operations on a list in Python.
#counting occurences of an element
my_list = [10, 20, 30, 20, 40, 20, 50]
occurrences = my_list.count(20)
print(f"Occurrences of 20: {occurrences}")
````

# --> TUPLE <--

````python
# Operations on tuples
my_tuple = (10, 20, 30, 40, 50)
print("Original tuple:", my_tuple)

# Accessing elements
print("First element:", my_tuple[0])
print("Last element:", my_tuple[-1])

# Slicing
print("First three elements:", my_tuple[:3])
print("Last three elements:", my_tuple[-3:])

# Length of tuple
print("Length of tuple:", len(my_tuple))

# Count occurrences
print("Count of 20 in tuple:", my_tuple.count(20))

# Index of an element
print("Index of 30 in tuple:", my_tuple.index(30))
# Note: Tuples are immutable, so we cannot add, remove, or modify elements.

# packing and unpacking
packed_tuple = 1, 2, 3, 4, 5
a, b, c, d, e = packed_tuple
print("Packed tuple:", packed_tuple)
print("Unpacked values:", a, b, c, d, e)



````

# --> DICTIONARY <--

````python
# Ordered collection of items
my_dict = {'Name': 'Alice', 'Age': 30, 'City': 'New York'}
print("Original dictionary:", my_dict)
# Accessing values
print("Name:", my_dict['Name'])
print("Age:", my_dict.get('Age'))       
# Adding a new key-value pair
my_dict['Profession'] = 'Engineer'
print("After adding Profession:", my_dict)
# Modifying an existing value
my_dict['Age'] = 31
print("After modifying Age:", my_dict)
# Removing a key-value pair
del my_dict['City']
print("After removing City:", my_dict)
# Iterating through keys and values
for key, value in my_dict.items():
    print(f"{key}: {value}")    
# Getting all keys
print("Keys:", my_dict.keys())
# Getting all values
print("Values:", my_dict.values())
# Clearing the dictionary
my_dict.clear()
print("After clearing:", my_dict)
# Note: Dictionaries are unordered collections prior to Python 3.7. From Python 3.7 onwards, they maintain insertion order.

````
