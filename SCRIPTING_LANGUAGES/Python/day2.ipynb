# Datatypes

You now know how to label the boxes (variables). Now we need to talk about exactly what goes inside them.

Python is smart. You don't have to tell it "This box holds a number." You just put the number in, and Python figures it out. This is called Dynamic Typing.

However, sometimes you need to know what type of data you are working with. For that, we use a special function called type().

We touched on the first three earlier, but now let's formalize them:

String (str): Text. Always inside quotes.

name = "Alice"

Integer (int): Whole numbers. No quotes.

age = 25

Float (float): Decimal numbers.

price = 19.99

Boolean (bool): True or False values. Logic gates. Important: Capitalize True and False.

is_game_over = False

```python
# How to check the type
# If you ever get confused about what is inside a variable, ask Python:

x = 10
y = "10"

print(type(x)) # Output: <class 'int'>
print(type(y)) # Output: <class 'str'>
```

The Magic of Re-assignment

In some languages (like C or Java), once a box holds a Number, it can only ever hold a Number. Python is flexible. You can dump the number out and put text in.

```python
my_box = 50       # Right now, it's an Integer
print(my_box)

my_box = "Hello"  # Now, it's a String!
print(my_box)
```

More About Datatypes -->

In Python, we categorize data types into two main families: Scalar Types (single values) and Collection Types (containers that hold multiple values). We also classify them by Mutability (can they be changed after creation?).

Here is your deep dive into the Python Data Type hierarchy.I. The Scalar Types (The Atoms)These represent a single piece of information.

1. Numeric Types -->

Integers (int): Whole numbers.Deep Knowledge: Unlike languages like C++ or Java, Python integers have arbitrary precision. This means they can be as big as your computer's memory allows. You can calculate $2^{1000}$ without an overflow error.

Floating Point (float): Decimal numbers.Deep Knowledge: These strictly follow the IEEE 754 standard (double precision). Be careful: computers struggle with precise decimal math (e.g., 0.1 + 0.2 might output 0.30000000000000004 due to binary representation limits)

.Complex Numbers (complex): Used in advanced engineering and math.Syntax: z = 3 + 5j. (Python uses j instead of i for the imaginary part).

2. Booleans (bool)

True / False: Used for logic.

Deep Knowledge: In Python, bool is actually a subclass of int.

True is mathematically equal to 1.

False is mathematically equal to 0.

Try this later: print(True + True) will output 2.

3. NoneType (None)

The concept of "Nothing": This is a special type representing the absence of a value. It is similar to null in other languages.

Usage: It is often used as a default value for variables that haven't been assigned yet.

II. The Sequence Types (Ordered Collections)

These types keep data in a specific order (1st, 2nd, 3rd...). You can access items by their index (position).

4. Strings (str)

Definition: A sequence of Unicode characters.

Deep Knowledge: Strings are Immutable. Once you create "Hello", you cannot change the 'H' to 'J'. You have to create a new string "Jello" to replace it.

Features:

III. The Mapping & Set Types (Unordered Collections)

Lets uderstand mutable and immutable-----

Mutable: The object can be changed after it is created. The box stays the same, but you can swap the contents.

Immutable: The object cannot be changed once created. If you want to change it, you have to destroy the old box and build a brand new one.

1. Mutable Types (The Changeable Ones)

Examples: list, dict, set

Imagine a List is like a Binder with pages in it.

You can open the binder, take a page out, write on a page, or add a new page.

The Binder itself (the variable) is still the same binder.

```python
# Lists are Mutable
my_list = [1, 2, 3]
print(id(my_list))  # id() prints the memory address (the box's location)

# Let's change the first item
my_list[0] = 99
print(my_list)      # Output: [99, 2, 3]
print(id(my_list))  # The ID is EXACTLY the same. Same box, different content.
```

2. Immutable Types (The Locked Ones)

Examples: int, float, str, tuple, bool

Imagine a String or Integer is like a Stone Tablet.

You carved "Hello" into the stone.

You cannot just erase the 'H' and write 'J'. You would have to get a new piece of stone and carve "Jello".

```python
# Strings are Immutable
word = "Hello"
# word[0] = "J"  <-- THIS WILL CRASH PYTHON (TypeError)
print(id(word))

# Instead, you must create a NEW string
word = "Jello"   # This isn't changing the old string; it's making a new one.
print(id(word))
```

```python
my_tuple = (10, 20)
# my_tuple[0] = 99  <-- ERROR! You cannot touch the data inside.
```

3. Why does this matter? (The Logic)

Performance: Immutable objects (like Tuples) are faster and lighter because Python knows they will never change size.

Safety: If you pass a List to a function, that function could delete your data by accident. If you pass a Tuple, your data is safe.

Dictionary Keys: This is advanced, but crucial. You can use a String or Tuple as a Key in a Dictionary because they don't change. You cannot use a List as a key because it might change, and Python would lose track of it.

-->and: Returns True only if BOTH sides are True.

-->or: Returns True if AT LEAST ONE side is True.

False or False -> False

-->not: Reverses the result.

not True -> False

5 . Membership Operators

Membership (in / not in)


```python
my_list = [1, 2, 3]
print(3 in my_list)      # True
print(99 not in my_list) # True
```

6. Identity Opeartors

Identity (is / is not)
This checks if two variables point to the same object in memory (not just if they look the same).

```python
a = [1, 2]
b = [1, 2]
print(a == b)  # True (Values are equal)
print(a is b)  # False! (They are two different boxes in memory)
```

7. Bitwise Operators (The Binary Architects)
Imagine we have two numbers:

a = 5 (In Binary: 0101)

b = 3 (In Binary: 0011)

Symbol,          |    Name,              |   Logic,                                |  Binary Action (5 & 3),     |        Result

  &,             |      Bitwise AND,     |     1 if both are 1,                    |      0101 & 0011 → 0001,    |         1
  
  |,             |      Bitwise OR,      |     1 if either is 1,                   |      0101 | 0011            |         1010
  
  ^,             |      Bitwise XOR,     |     1 if bits are different,            |      0101 ^ 0011 → 0110,    |         6
  
  ~,             |      Bitwise NOT,     |     "Flips 0 to 1, 1 to 0",             |      ~0101 → 1010,          |        -6*
  
  <<,            |      Left Shift,      |     Pushes bits left (Multiply by 2),   |      0101 << 1 → 1010,      |         10
  
  >>,            |      Right Shift,     |     Pushes bits right (Divide by 2),    |      0101 >> 1 → 0010,      |          2
  




```python
# Bonus: The "New" Operator (Python 3.8+)
# There is one more operator that was added recently, often called the Walrus Operator because it looks like eyes and tusks: :=

# It allows you to assign values inside an expression.

# Without Walrus:

a = len("Hello")
if a > 3:
    print(f"Too long! Length is {a}")


# With Walrus:

if (n := len("Hello")) > 3:
    print(f"Too long! Length is {n}")

```
