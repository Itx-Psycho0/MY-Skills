# Variables, Datatypes and Operators


--> Imagine you are moving house. You wouldn't just throw all your items into the moving truck loosely, right? You would put them in boxes and label them.

In Python, Variables are those boxes.

The Box: The variable itself.

The Item inside: The data (String, Integer, Float, etc.).

The Label: The variable name.

```python
# How to create a variable 

# syntax: variable_name = value

player_name = "Mario"   # A box labeled 'player_name' containing the string "Mario"
score = 1500            # A box labeled 'score' containing the integer 1500
health = 98.5           # A box labeled 'health' containing the float 98.5
```

Why do we use them?

Variables allow us to store data and use it later. Instead of typing "Mario" every time, we just type player_name.

The Rules of Naming (Syntax)

Python has strict rules about how you can label your boxes. If you break them, you get a SyntaxError.

Must start with a letter or underscore _. (Cannot start with a number!)

✅ user1

❌ 1user

Can only contain letters, numbers, and underscores. (No spaces or special symbols like $, @, !)

✅ my_variable

❌ my variable (Space is illegal)

❌ money$

Case Sensitive.

# Datatypes

You now know how to label the boxes (variables). Now we need to talk about exactly what goes inside them.

Python is smart. You don't have to tell it "This box holds a number." You just put the number in, and Python figures it out. This is called Dynamic Typing.

However, sometimes you need to know what type of data you are working with. For that, we use a special function called type().

We touched on the first three earlier, but now let's formalize them:

String (str): Text. Always inside quotes.

name = "Alice"

Integer (int): Whole numbers. No quotes.

age = 25

Float (float): Decimal numbers.

price = 19.99

Boolean (bool): True or False values. Logic gates. Important: Capitalize True and False.

is_game_over = False

```python
# How to check the type
# If you ever get confused about what is inside a variable, ask Python:

x = 10
y = "10"

print(type(x)) # Output: <class 'int'>
print(type(y)) # Output: <class 'str'>
```

The Magic of Re-assignment

In some languages (like C or Java), once a box holds a Number, it can only ever hold a Number. Python is flexible. You can dump the number out and put text in.

```python
my_box = 50       # Right now, it's an Integer
print(my_box)

my_box = "Hello"  # Now, it's a String!
print(my_box)
```
