```python
import seaborn as sns

import matplotlib.pyplot as plt

d1=pd.DataFrame(
    {'Student':['A','B','C','D'],
     'Marks':[90,80,40,60]
    })

sns.boxplot(data=d1, x= 'Student', y='Marks', showmeans=True)

plt.show()
```

```python
pivot = d1.pivot_table(index='Student', values='Marks')
sns.heatmap(pivot, annot=True, cmap='YlGnBu')
plt.show()
```

```python
sns.pairplot(d1, hue='Student')
plt.show()
```
