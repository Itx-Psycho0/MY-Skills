# Loops

1. The while Loop (The "Repeating If")
Think of a while loop exactly like an if statement, but it doesn't stop after running once. It keeps running as long as the condition is True.

syntax:-

while condition:

    # Code to repeat

    # IMPORTANT: You must change something here to eventually make the condition False
    

    
The Logic Trap (Infinite Loops): If the condition never becomes False, your program runs forever until it crashes your computer's memory. This is the most common bug for beginners.

```python
#Example:
battery = 10  # Start with 10% battery

while battery > 0:
    print(f"Phone is on. Battery: {battery}%")
    battery = battery - 1  # Crucial step: We drain the battery
    
print("Phone died.")
```

2. The for Loop (The "Sequence Master")

This is the most popular loop in Python. Unlike C or Java (which use i = 0; i < 10; i++), Python's for loop is designed to iterate over a collection (like a List or a Range of numbers).

Syntax:


for variable_name in collection:

    # Do something with variable_name

The range() Function: When you just want to loop a specific number of times, you use range(start, stop, step).

range(5) → 0, 1, 2, 3, 4 (Stops before the number!)


range(1, 10, 2) → 1, 3, 5, 7, 9 (Step size of 2)

```python
# Counting down logic
for number in range(10, 0, -1):
    print(f"T-minus {number}")
print("Blastoff!")
```

3. Controlling the Flow (break and continue)

You are the master of the loop. You can stop it or skip parts of it using these keywords:

break: "Emergency Stop." Kills the loop immediately and moves to the code after it.

Use Case: You are searching for a specific user. Once you find them, stop looking.

continue: "Skip." Stops the current round and jumps back to the top for the next round.

Use Case: You are processing payments. One payment fails? continue to the next one; don't stop the whole system.

```python
for floor in range(1, 11): # Floors 1 to 10
    if floor == 4:
        continue  # Superstition! Skip floor 4.
    
    if floor == 8:
        print("Penthouse reached. Stopping elevator.")
        break     # Stop going up.
        
    print(f"Stopping at Floor {floor}")
```

4. The "Else" Block in Loops (Deep Knowledge)

This is a rare Python feature that most tutorials miss. You can attach an else block to a loop.

Logic: The else code runs ONLY if the loop finished naturally (without hitting a break).

```python
# Search Logic
target = 5
numbers = [1, 2, 3]

for n in numbers:
    if n == target:
        print("Found it!")
        break
else:
    # This runs because we never broke out of the loop
    print("Number not found in list.")
```

```python
# Initialize countdown value
count = 10  

# While loop for countdown
while count > 0:
    print(count)
    count -= 1  # Decrease the count

# Print liftoff message after countdown
print("Liftoff!")  
```

```python
# Initialize the roll number
roll_number = 1  

# Loop runs while roll number is less than or equal to 10
while roll_number <= 10:
    print("Roll Number:", roll_number)  # Print the current roll number
    roll_number += 2  # Increment by 2, to print alternate roll numbers

print("All roll numbers printed!")  # Executes after the loop ends
```
