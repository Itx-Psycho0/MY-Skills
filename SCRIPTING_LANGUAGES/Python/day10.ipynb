# Numpy

```python
import numpy as np

# Arithmetic opr
a1 = np.array([1,2,3,4,5])
a2 = np.array([6,7,8,9,10])

a1 + a2
a1 - a2
a1 * a2
a1 // a2
a1 ** a2
```

```python
# broadcasting
a1 +10
```

```python
# Deep and Shallow copy

# shallow copy
a1
slice = a1[:3]
slice = slice * 10
slice
a1
b1 = a1.copy()# makes shallow copy of it
# so there is no change in original array but if we do in list they changed.

# deep copy
b1 = a1
b1[0]=99
a1
```

```python
# Matrix opr
m1 = np.array([[1,2],[3,4]])
m2 = np.array([[5,6],[7,8]])

m1
```

```python
m2
```

```python
# multiplication
m1 @ m2
np.dot(m1,m2)
```

```python
# transpose
m1.T
```

```python
# Advance Array Manipulation

# Stacking of array
#you can't stack two matrix

np.vstack((a1,a2)) # vertical stack
np.hstack((a1,a2)) # horizontal stack
np.column_stack((a1,a2)) # column stack

```
