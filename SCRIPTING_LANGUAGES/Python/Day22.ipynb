```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
data = pd.DataFrame({
    'Color': ['Red', 'Green', 'Blue', 'Green', 'Red']   
})

data
```

```python
le = LabelEncoder()
data['Color_Encoder'] = le.fit_transform(data['Color'])

data
```

```python
Encoded_data = pd.get_dummies(data,columns=['Color'])
Encoded_data
```
