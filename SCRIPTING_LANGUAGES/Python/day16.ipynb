# Matrix plot and regression plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
flights = sns.load_dataset('flights')
flights
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
tipscorr = tips[['total_bill','tip','size']]
```

```python
tipscorr.corr()
```

```python
sns.heatmap(tips.corr())
```

```python
sns.clustermap(tipscorr.corr())
```

```python
pvtFlight= flights.pivot_table(values= 'passengers',index = 'month',columns= 'year')
pvtFlight
```

```python
sns.heatmap(pvtFlight)
```

```python
# Regression plot
# regression plot is used in ml so we will not fully discuss what, how and where it is used but we will see how it is created and what are the values we cna put inside it .
```

```python
# use Lmplot()
tips = sns.load_dataset("tips")
tips
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips)
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, hue='sex')
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips,hue='sex', markers=['o','v'])
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips,hue='sex', markers=['o','v'],scatter_kws={'s':100})
```
