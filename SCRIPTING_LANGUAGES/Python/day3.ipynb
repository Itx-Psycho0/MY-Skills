# Topic :- Control Flow Statements in python


In Python, we use Conditional Statements. Unlike other languages that use curly braces {}, Python uses Indentation (spaces) to show which code belongs to which decision.

```python
# The Syntax Structure of Python Programming Language
#if condition:
    # Do this if True
#elif another_condition:
    # Do this if the first was False, but this is True
#else:
    # Do this if EVERYTHING above was False
```

When writing logic, always ask: "What are the boundaries?" For a number range, the boundaries are usually the specific numbers where the behavior changes.

Example: If a passing grade is 50:

49 is a Fail.

51 is a Pass.

50 is the Edge Case. (Is it pass or fail? You must decide explicitly using >= or >).

```python
# Sometimes a decision depends on a previous decision. You can put an if block inside another if block. This is called Nesting.

# Concept: Think of it like security doors. You must pass Door 1 to even get to Door 2.
# Example of Nested If Statements in Python

username = "admin"
password = "123"
is_account_locked = False

if username == "admin":
    # We only check password if username matches
    if not is_account_locked:
        if password == "123":
            print("Access Granted")
        else:
            print("Wrong Password")
    else:
        print("Account Locked")
else:
    print("Unknown User")
# Warning: Too much nesting (like 5 levels deep) is called "Spaghetti Code." It is messy. We usually try to flatten this logic using and/or operators, but you must understand nesting to read other people's code.
```

```python
# The "Switch Case" (Python's match)

# For a long time, Python did not have a switch case. Programmers had to use long if-elif-else chains. But in Python 3.10, they added Structural Pattern Matching.

# This is the keyword match. It is Python's super-powered version of Switch Case.

# Syntax:

status_code = 404

match status_code:
    case 200:
        print("Success")
    case 400:
        print("Bad Request")
    case 404:
        print("Not Found")
    case 500 | 501:  # The | symbol acts like "OR"
        print("Server Error")
    case _:          # The underscore (_) acts as the "Default" or "Else"
        print("Unknown Status")
# Why is this better than elif?

# Cleanliness: It is much easier to read when checking one variable against many values.

# Pattern Matching: You can match complex data structures (like lists), not just simple numbers.
```

```python
# The Ternary Operator (One-Line Decision)
# This is the tool for experts who want concise code. It allows you to condense 4 lines of if-else into 1 line.

# Standard Way:

age = 20
if age >= 18:
    status = "Adult"
else:
    status = "Child"

# Ternary Way:
# Syntax: value_if_true if condition else value_if_false
status = "Adult" if age >= 18 else "Child"
```
