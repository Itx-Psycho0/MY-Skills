# HASHING

- Hashing is a technique used in data structures that efficiently stores and retrieves data in a way that allows for quick access.


```python
# TWO SUM
def twoSum(nums,target):
    n = len(nums)
    dict = {}
    for i in range(n):
        rem = target - nums[i]
        if rem in dict:
            return [dict[rem],i]
        dict[nums[i]] = i

print(twoSum([2,7,11,15],9))


 
```

```python
# twoSum second method
def twoSum(numbers,target):
    left,right = 0,len(numbers)-1
    while left<right:
        current_sum= numbers[left]+numbers[right]
        if current_sum == target:
            return[left+1,right+1]
        elif current_sum < target:
            left+=1
        else:
            right -=1


print(twoSum([2,7,11,15],9))
```

```python
# intersection of array
def intersection(l1,l2):
    return list(set(l1)&set(l2))

print(intersection([1,2,2,1],[2,2]))
```

```python
# first unique character
def firstUniqueChar(s):
    freq = {}
    for i in s:
        if i not in freq:
            freq[i]=1
        else:
            freq[i] +=1
    for i in range(len(s)):
        if freq[s[i]]==1:
            return i
    return -1

print(firstUniqueChar("loveleetcode"))


```

```python
# ANAGRAM
def isAnagram(s,t):
    freq={}
    if len(s)!=len(t):
        return False
    for i in s:
        if i not in freq:
            freq[i]=1
        else:
            freq[i]+=1
    for i in t:
        if i not in freq:
            return False
        else:
            freq[i]-=1

    for i in freq.values():
        if i!=0:
            return False
    return True

print(isAnagram("anagram","nagaram"))
        
```

```python
# group anagram:
def sortingstring(s):
    s1 = list(s)
    s1.sort()
    return "".join(s1)
def groupAnagram(strs):
    freq={}
    for i in strs:
        key = sortingstring(i)
        if key in freq:
            freq[key].append(i)
        else:
            freq[key]= [i]
    return list(freq.values())


print(groupAnagram(["eat","tea","tan","nat","ate","bat"]))
```
