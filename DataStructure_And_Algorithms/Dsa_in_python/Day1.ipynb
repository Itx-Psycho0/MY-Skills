-> What is a Programming Language?

A programming language is a formal set of instructions that can be used to produce various kinds of output, such as software applications, websites, and scripts. It provides a way for humans to communicate with computers and instruct them to perform specific tasks.

-> High-level vs Low-level Programming Languages

:- High-level programming languages are designed to be easy for humans to read and write. They use natural language elements and abstract away much of the complexity of the underlying hardware. Examples include Python, Java, and C++.

:- Low-level programming languages are closer to machine code and provide little abstraction from the hardware. They are more difficult for humans to read and write but offer greater control over the computer's resources. Examples include Assembly and machine code.

-> Compiled vs Interpreted Languages

:- Compiled languages are translated into machine code by a compiler before they are executed. This results in faster execution times since the code is already in a format that the computer can understand. Examples include C, C++, and Rust.

:- Interpreted languages are executed line-by-line by an interpreter without being compiled into machine code first. This can make them slower but allows for more flexibility and easier debugging. Examples include Python, JavaScript, and Ruby.


---------------------------------------------------

-> Compiler vs Interpreter

:- A compiler is a program that translates the entire source code of a programming language into machine code before execution. It checks for errors and optimizes the code for performance. Once compiled, the resulting machine code can be executed directly by the computer's hardware.

:- An interpreter, on the other hand, translates and executes the source code line-by-line at runtime. It does not produce a separate machine code file, and it can be more flexible for testing and debugging since changes can be made and executed immediately without recompilation.


-> What is Python?
Python is a high-level, interpreted programming language known for its simplicity and readability. It was created by Guido van Rossum and first released in 1991. Python supports multiple programming paradigms, including procedural, object-oriented, and functional programming. It has a large standard library and a vibrant ecosystem of third-party packages, making it suitable for a wide range of applications, from web development to data science and artificial intelligence.

-> Why Python?
:- Easy to Learn and Read: Python's syntax is clear and straightforward, making it accessible for beginners.
:- Versatile: Python can be used for web development, data analysis, machine learning, automation, and more.
:- Large Community: Python has a vast and active community, providing extensive resources, libraries, and frameworks.
:- Cross-Platform: Python runs on various operating systems, including Windows, macOS, and Linux.
:- Strong Support for Integration: Python can easily integrate with other languages and technologies, making it a flexible choice for various projects.

```python
# Control Statements
#if-else example:-
num = int(input("Enter a number: "))
if num % 2 == 0:
    print(num, "is even")
else:
    print(num, "is odd")

#ternary operator example:-
age = int(input("Enter your age: "))
status = "Adult" if age >= 18 else "Minor"
print("You are an", status)
```
