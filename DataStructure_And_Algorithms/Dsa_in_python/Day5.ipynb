# Bubble Sort



```python
arr = [3,2,8,1,7,6,4]
n = len(arr)
for i in range(n):
    isSwap = False
    for j in range(n-i-1):
        if arr[j]>arr[j+1]:
            arr[j],arr[j+1]=arr[j+1],arr[j]
            isSwap = True
    if not isSwap:
        break
print(arr)


# Space Complexity = O(1)
# Time Complexity = O(n^2) in wrost case and O(n) in best case
```

# Insertion Sort

```python
arr = [3,2,8,1,7,6,4]
for i in range(1,n):
    key = arr[i]
    j = i-1
    while j>=0 and arr[j]>key:
        arr[j+1] = arr[j]
        j -=1
    arr[j+1]=key
print(arr)


# Space Complexity = O(1)
# Time Complexity = O(n^2) in wrost case and O(n) in best case
```

# Selection Sort

```python
arr = [3,2,8,1,7,6,4]

for i in range(len(arr)):
    mn = arr[i]
    ind = i
    for j in range(i+1,len(arr)):
        if arr[j]<mn:
            mn = arr[j]
            ind = j
    arr[i],arr[ind]=arr[ind],arr[i]
print(arr)

# Space Complexity = O(1)
# Time Complexity = O(n^2) in wrost case and O(n^2) in best case        
```
