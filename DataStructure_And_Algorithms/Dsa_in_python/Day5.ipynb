# Bubble Sort



```python
arr = [3,2,8,1,7,6,4]
n = len(arr)
for i in range(n):
    isSwap = False
    for j in range(n-i-1):
        if arr[j]>arr[j+1]:
            arr[j],arr[j+1]=arr[j+1],arr[j]
            isSwap = True
    if not isSwap:
        break
print(arr)


# Space Complexity = O(1)
# Time Complexity = O(n^2) in wrost case and O(n) in best case
```

# Insertion Sort

```python
arr = [3,2,8,1,7,6,4]
for i in range(1,n):
    key = arr[i]
    j = i-1
    while j>=0 and arr[j]>key:
        arr[j+1] = arr[j]
        j -=1
    arr[j+1]=key
print(arr)


# Space Complexity = O(1)
# Time Complexity = O(n^2) in wrost case and O(n) in best case
```

# Selection Sort

```python
arr = [3,2,8,1,7,6,4]

for i in range(len(arr)):
    mn = arr[i]
    ind = i
    for j in range(i+1,len(arr)):
        if arr[j]<mn:
            mn = arr[j]
            ind = j
    arr[i],arr[ind]=arr[ind],arr[i]
print(arr)

# Space Complexity = O(1)
# Time Complexity = O(n^2) in wrost case and O(n^2) in best case        
```

```python
# Sort Colors

nums=[1,0,2,2,0,1,1,0,2,0]
left = 0
right = len(nums)-1
i = 0
while i<=right:
    if nums[i]==1:
        i+=1
    elif nums[i]==0:
        nums[i],nums[left]=nums[left],nums[i]
        i+=1 
        left +=1
    else:
        nums[i],nums[right]=nums[right],nums[i]
        right -=1
        
print(nums)
    
```

# Linear search

```python
nums = [1,2,4,5,9,7,6]
target = 9

for i in range(0,len(nums)):
    if nums[i]==target:
        print(i)

# Time complexity = O(n)
```

# Binary Search

```python
def binary_search(arr, target):
    low = 0
    high = len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid -1
    return -1
print(binary_search([1,2,3,4,5,6],5))
```
