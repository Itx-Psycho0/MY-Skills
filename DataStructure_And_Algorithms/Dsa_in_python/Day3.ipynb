```python
# PRACTICE PROBLEMS

#method 1
def prefixSum(arr):
    sum = [0]*len(arr)
    for i in range(0,len(arr)):
        sum[i]= arr[i] + sum[i-1]
    return sum

print(prefixSum([3,1,8,2,10,5]))


# method 2

def prefixSum2(arr):
    for i in range(1,len(arr)):
        arr[i]=arr[i-1]+arr[i]
    return arr

print(prefixSum([3,1,8,2,10,5]))
```

```python
def removeDuplicatedFromSortedArray(nums):
    start = 0
    for i in range(1,len(nums)):
        if nums[i] != nums[start]:
            start +=1
            nums[start]=nums[i]
    return start+1

print(removeDuplicatedFromSortedArray([3,3,5,7,7,7,9,9,9])) 


```

```python
def removeDuplictes2(nums):
    if len(nums)<=2:
        return len(nums)
    
    start =1 
    for i in range(2,len(nums)):
        if nums[i]!= nums[start-1]:
            start +=1
            nums[start] = nums[i]
    return start+1

print(removeDuplictes2([1,1,1,2,2,3]))
```

```python
def sortArrayParity(nums):
    n = len(nums)
    start = 0
    for i in range(0,n):
        if nums[i]%2==0:
            temp = nums[i]
            nums[i]=nums[start]
            nums[start]=temp
            start +=1
    return nums


print(sortArrayParity([3,1,2,4]))
```

# Kadane's Algorithm

use to find maximum subarrray

in all positive array maximum array will be full array

but in all negative number array maximum array will be greatest value in the array like -1

```python
def maxSubArrray(nums):
    current = 0
    maxSum = nums[0]
    for i in range(len(nums)):
        current += nums[i]
        if current>maxSum:
            maxSum = current
        if current <0:
            current = 0
    return maxSum

print(maxSubArrray([-2,1,-3,4,-1,2,1,-5,4]))
```

```python
def maxProfit(nums):
    mini_price = nums[0]
    profit = 0
    for i in range(1,len(nums)):
        curr_profit = nums[i]-mini_price
        if curr_profit>profit:
            profit = curr_profit
        mini_price = min(mini_price,nums[i])
    return profit

print(maxProfit([7,1,5,3,6,4]))
```

```python
def maxProfit2(arr):
    profit = 0
    n = len(arr)
    for i in range(n-1):
        if arr[i+1]>arr[i]:
            profit += arr[i+1]-arr[i]
    return profit

print(maxProfit2([7,1,5,3,6,4]))
```

```python
def richestWealthMax(arr):
    ans = 0
    for acc in arr:
        ans = max(ans,sum(acc))
    return ans

print(richestWealthMax([[2,3],[5,6]]))
```

```python
def defangIPAddr(address):
    ans = ""
    for i in address:
        if i != ".":
            ans +=i
        else :
            ans += "[.]"
    return ans

print(defangIPAddr("255.100.50.0"))
```
