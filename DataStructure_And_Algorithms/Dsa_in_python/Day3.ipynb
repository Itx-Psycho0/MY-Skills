# Lists, Tuples, Strings

--> what is List ?

So list is a data structure data stores data in sequence and order manner, the data is homogeneous or heterogeneous doesn't matter. it is mutable, ordered and allow duplicates.

--> diff b/w List and array:-

in other languages array is fixed , homogeneous data structure , but the list is not in python , it also work as array or not.



list can be accessing , modifying etc.


--> Operations on List :-

1. len() // find the length of list
2. min() // find the minimum value in the list
3. max() // find the maximum value in the list
4. negative indexing // we can do access list by negative numbers also
5. append() // add the vlaue in the last 
6. extend() // extend the list by other list
7. insert() // insert the value on any index in the list
8. remove() // remove the specific value form the list
9. pop() // remove the value from the last and also on given index
10. clear() // clear the list
11. slicing // slice the value at given index range
12. count() // count the given value from the list
13. sort() // sort the list in increasing order
14. reverse() // reverse the list 
15. copy() // Copy the list
16. index() // given the index of value
