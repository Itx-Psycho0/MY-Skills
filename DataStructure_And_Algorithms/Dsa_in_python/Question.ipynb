```python
# Power of two
def powerOfTwo(n):
    if n<=0:
        return False
    while n>1:
        if n%2!=0:
            return False
        n //=2
    return True

print(powerOfTwo(16))
```

```python
# MIN AND MAX IN ARRAY
def minAndMax(arr):
    mn=arr[0]
    mx=arr[0]
    for i in range(len(arr)):
        if mx < arr[i]:
            mx = arr[i]
        elif mn>arr[i]:
            mn = arr[i]
    return (mn,mx)

print(minAndMax([5, 2, 8, 1, 10, 3]))
```

```python
# Reverse an Array
def reverseArray(arr):
    left = 0
    right = len(arr)-1
    while left<right:
        arr[left],arr[right]=arr[right],arr[left]
        left +=1
        right-=1
    return arr

print(reverseArray( [1, 4, 3, 2, 6, 5]))
```
