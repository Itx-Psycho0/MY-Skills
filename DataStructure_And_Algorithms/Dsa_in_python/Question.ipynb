```python
# Power of two
def powerOfTwo(n):
    if n<=0:
        return False
    while n>1:
        if n%2!=0:
            return False
        n //=2
    return True

print(powerOfTwo(16))
```

```python
# MIN AND MAX IN ARRAY
def minAndMax(arr):
    mn=arr[0]
    mx=arr[0]
    for i in range(len(arr)):
        if mx < arr[i]:
            mx = arr[i]
        elif mn>arr[i]:
            mn = arr[i]
    return (mn,mx)

print(minAndMax([5, 2, 8, 1, 10, 3]))
```

```python
# Reverse an Array
def reverseArray(arr):
    left = 0
    right = len(arr)-1
    while left<right:
        arr[left],arr[right]=arr[right],arr[left]
        left +=1
        right-=1
    return arr

print(reverseArray( [1, 4, 3, 2, 6, 5]))
```

```python
# Find Duplicates
def findDuplicates(arr):
    # seen = set()
    # res = []
    # for num in arr:
    #     if num not in seen:
    #         seen.add(num)
    #     else:
    #         res.append(num)
    # return res
    res=[]
    freq ={}
    for num in arr:
        if num not in freq:
            freq[num]=1
        else:
            freq[num]+=1
    for key in freq.keys():
        if freq[key] > 1:
            res.append(key)
    return res

print(findDuplicates([4, 3, 2, 7, 8, 2, 3, 1]))
```

```python
# intersection of two array
def intersection(arr1,arr2):
    s1= set(arr1)
    s2 = set(arr2)
    return list(s1&s2)

print(intersection([1,3,4,5],[3,5,6,8]))
```

```python
# Running sum of 1d array
def runningSum(arr):
    sum=0
    for i in range(len(arr)):
        sum += arr[i]
        arr[i]=sum
    return arr

print(runningSum([1, 2, 3, 4]))
```

```python
# sorted Square
def sortedSquares(arr):
    # return sorted([n**2 for n in arr])
    for n in range(len(arr)):
        arr[n] **=2
    arr.sort()
    return arr

print(sortedSquares([-4,-1,0,3,10]))
```

```python
# unique occurence
def uniqueOcuurence(arr):
    freq = {}
    for i in arr:
        if i not in freq:
            freq[i]=1
        else:
            freq[i]+=1
    count = list(freq.values())
    return len(freq) == len (set(count))

print(uniqueOcuurence([1,2,2,1,1,3]))
```

```python
# palindrome
def palindrome(n):
    if n<0:
        return False
    rev =0
    while n>0:
        rev *=10 + n%10
        n//=10
    return rev == n

print(palindrome(121))

```

```python
# Subarray sum equals k
def subarraySum(nums, k):
    prefix_count = {0: 1}  # important
    curr_sum = 0
    count = 0

    for num in nums:
        curr_sum += num
        if curr_sum - k in prefix_count:
            count += prefix_count[curr_sum - k]
        prefix_count[curr_sum] = prefix_count.get(curr_sum, 0) + 1

    return count

print(subarraySum([1, 2, 3],3))
```
