# Linked List

```python
# Create a linked list.
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

n1 = Node(10)
n2 = Node(20)
n3 = Node(30)
n1.next = n2
n2.next = n3
print(n1.data)  # Output: 10
print(n1.next.data)  # Output: 20
print(n1.next.next.data)  # Output: 30
```

```python
class LinkedList:
    def __init__(self):
        self.head = None
    def append(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        last = self.head
        while last.next:
            last = last.next
        last.next = new_node

    def quick_print(self):
        current = self.head
        while current:
            print(current.data, end=" -> ")
            current = current.next
        print("None")
ll = LinkedList()
ll.append(10)
ll.append(20)
print(ll.head.data)       
ll.quick_print()          
```

```python
# Fast and Slow Pointer Technique
def find_middle(head):
    slow = head
    fast = head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
    return slow.data
print(find_middle(ll.head))  # Output: 20
```

```python
# cycle detection
def has_cycle(head):
    slow = head
    fast = head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        if slow == fast:
            return True
    return False    
print(has_cycle(ll.head))  # Output: False
```
