```python
# second largest element
# 1st method
arr = [3, 1, 4, 1, 5, 9]
n = len(arr)
arr.sort() # n log n
second_largest = arr[-2]
print("Second largest element:", second_largest)
```

```python
# 2nd method
arr = [3, 1, 4, 1, 5, 9]
n = len(arr)
max_el = 0
for i in range(n):
    max_el = max(max_el, arr[i])
second_largest = -1
for i in range(n):
    if arr[i] != max_el:
        second_largest = max(second_largest, arr[i])
print("Second largest element:", second_largest)
```

```python
# 3rd method
arr = [3, 1, 4, 1, 5, 9]
n = len(arr)
max_el = arr[0]
second_largest = float('-inf')
for i in range(1, n):
    if arr[i] > max_el:
        second_largest = max_el
        max_el = arr[i]
    # elif arr[i] != max_el:
    #     second_largest = max(second_largest, arr[i])
print("Second largest element:", second_largest)

# time complexity of 2nd and 3rd method is O(n)
```

```python
# find the smallest subarray  with sum >= target
arr = [2,3,1,2,4]
target = 7
n = len(arr)
left = 0
current_sum = 0
min_length = float('inf')
for right in range(n):
    current_sum += arr[right]
    while current_sum >= target:
        min_length = min(min_length, right - left + 1)
        current_sum -= arr[left]
        left += 1
if min_length == float('inf'):
    print(0)
else:
    print(min_length)
```
